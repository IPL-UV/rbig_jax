## Classic RBIG Dataset

```python
n_samples = 2_000
n_features = 2

data = get_classic(n_samples)

sns.jointplot(data[:, 0], data[:, 1], s=5, color='red')
```

## Noisy Circles

```python
from sklearn import cluster, datasets

data, y = datasets.make_circles(
    n_samples=n_samples, factor=.5,
    noise=.05, random_state=123)


sns.jointplot(data[:, 0], data[:, 1], s=5, color='red')
```

## Moons Dataset

```python
data, y = datasets.make_moons(
    n_samples=n_samples,
    noise=.05, random_state=123)


sns.jointplot(data[:, 0], data[:, 1], s=5, color='red')
```

## Swiss Roll

```python
data, y = datasets.make_s_curve(n_samples=n_samples,noise=0.1, random_state=123)

sns.jointplot(data[:, 0], data[:, 2], s=5, color='red')
```

## Double Helix

```python
data, y = datasets.make_swiss_roll(n_samples=n_samples,noise=0.5, random_state=123)

sns.jointplot(data[:, 0], data[:, 2], s=5, color='red')
```

## Blobs

```python
data, y = datasets.make_blobs(n_samples=n_samples, random_state=3, n_features=2,)

sns.jointplot(data[:, 0], data[:, 1], s=5, color='red')
```
